```python
# ðŸ“˜ 1_Data_Preprocessing.ipynb
import pandas as pd
import os
```

```python
# Load raw dataset
df = pd.read_csv("../data/raw/india_districts_2500.csv")
print("âœ… Data loaded:", df.shape)
```

```python
# Melt monthly rainfall columns into one column
month_cols = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
df_long = df.melt(
    id_vars=['DISTRICT', 'YEAR', 'AVG_TEMP', 'HUMIDITY', 'ELEVATION', 'FLOODS'],
    value_vars=month_cols,
    var_name='MONTH',
    value_name='RAINFALL_MM'
)
```

```python
# Encode FLOODS as binary
df_long['FLOODS'] = df_long['FLOODS'].map({'YES': 1, 'NO': 0})
```

```python
# Map months to numbers
month_map = {m: i+1 for i, m in enumerate(month_cols)}
df_long['MONTH'] = df_long['MONTH'].map(month_map)
```

```python
# Drop missing values
df_long.dropna(inplace=True)
```

```python
# Save processed data
os.makedirs("../data/processed", exist_ok=True)
df_long.to_csv("../data/processed/flood_cleaned.csv", index=False)
```
